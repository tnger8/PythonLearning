```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('potatoe.jpg')
```

```python
plt.imshow(img) # BGR
```

```python
img_rgb= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Filp the image upside down 
new_img = img_rgb.copy()
```

```python
new_img = cv2.flip(new_img, 0)
plt.imshow(new_img)
```

```python
# draw red rectangle ard the centre of image 

pt1 = (400,200)
pt2 =  (100, 500)
```

```python
cv2.rectangle(img_rgb, pt1=pt1, pt2=pt2, color=(0,255,0), thickness=10)
plt.imshow(img_rgb)
```

```python
# draw triangle at the centre
vertices = np.array([[600,300], [800,300], [700,200]], np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1, 1 ,2))
```

```python
pts.shape
```

```python
cv2.polylines(img_rgb, [pts], isClosed= True, color = (0,0,255), thickness=11)
plt.imshow(img_rgb)
```

```python
import cv2
import numpy as np

#################################
## Function ###
#################################

def draw_circle(event, x, y, flags, param):
    
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y), radius=100, color=(0,55,0), thickness=10)

img = cv2.imread("potatoe.jpg")
cv2.namedWindow(winname="potatsoe")
cv2.setMouseCallback("potatsoe", draw_circle)


while True:
    cv2.imshow('potatsoe', img)
    if cv2.waitKey(20) & 0xFF ==27:
        break
        
cv2.destroyAllWindows()
```
